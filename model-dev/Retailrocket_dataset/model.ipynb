```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Embedding, concatenate, Input
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

import os
import joblib
import pandas as pd
import numpy as np

import json
from tqdm import tqdm
# 创建存储目录
os.makedirs('./preprocessed_data', exist_ok=True)

# %% [markdown]
# ## 1. 数据加载与预处理
print("✅ 开始执行：Retailrocket 用户行为预测系统")

print("🔄 步骤1：数据加载")
events = pd.read_csv('events.csv')
print("✅ 数据加载完成")

print("🔄 步骤2：数据清洗")
events = events.dropna(subset=['visitorid', 'event'])
events['timestamp'] = pd.to_datetime(events['timestamp'], unit='ms')
print("✅ 数据清洗完成")

# %% [markdown]
# ## 修正后的完整流程

print("🔄 步骤3：生成高危因子特征")
# 生成用户级特征（每个用户一行）
# 在步骤3前添加高危因子生成函数
def create_risk_features(df):
    user_agg = df.groupby('visitorid').agg(
        total_events=('event', 'count'),
        cart_additions=('event', lambda x: (x == 'addtocart').sum()),
        purchases=('event', lambda x: (x == 'transaction').sum())
    ).reset_index()
    
    user_agg['cart_abandon_rate'] = np.where(
        user_agg['cart_additions'] > 0,
        (user_agg['cart_additions'] - user_agg['purchases']) / user_agg['cart_additions'],
        0  # 处理除零情况
    )
    
    last_active = df.groupby('visitorid')['timestamp'].max().reset_index()
    return user_agg.merge(last_active, on='visitorid')
# 生成特征后保存
risk_features_path = './preprocessed_data/risk_features.parquet'

# 检查是否已有缓存
if not os.path.exists(risk_features_path):
    risk_features = create_risk_features(events)
    # 保存为Parquet格式（轻量高效）
    risk_features.to_parquet(risk_features_path)
    print("✅ 高危因子特征已保存")
else:
    # 直接加载已有数据
    risk_features = pd.read_parquet(risk_features_path)
    print("✅ 从缓存加载高危因子特征")
print(f"✅ 高危因子特征生成完成 | 用户数: {len(risk_features)}")

print("🔄 步骤4：构造时间序列数据（带用户ID）")

encoder = LabelEncoder()
encoder.fit(['view', 'addtocart', 'transaction'])

def create_sequences(df, seq_length=10):
    df = df.sort_values(['visitorid', 'timestamp'])
    
    sequences = []  # 初始化序列列表
    targets = []    # 初始化目标列表
    user_ids = []   # 初始化用户ID列表
    
    # 使用tqdm显示进度（需提前导入库）
    from tqdm import tqdm
    grouped = tqdm(df.groupby('visitorid'), desc="生成序列", unit="用户")
    
    for uid, group in grouped:
        events = encoder.transform(group['event'])  # 编码事件类型
        # 滑动窗口生成序列
        for i in range(len(events) - seq_length):
            sequences.append(events[i:i+seq_length])
            targets.append(events[i+seq_length])
            user_ids.append(uid)
    
    return np.array(sequences), np.array(targets), np.array(user_ids)

# 生成序列后保存
seq_data_path = './preprocessed_data/seq_data.npz'

# 参数化序列长度（便于后续调整参数）
SEQ_LENGTH = 10

# 检查是否已有缓存
if not os.path.exists(seq_data_path):
    X_seq, y, sequence_user_ids = create_sequences(events, seq_length=SEQ_LENGTH)
    # 保存为压缩的numpy格式
    np.savez_compressed(
        seq_data_path,
        X_seq=X_seq,
        y=y,
        sequence_user_ids=sequence_user_ids
    )
    print(f"✅ 时间序列数据已保存（序列长度={SEQ_LENGTH}）")
else:
    # 加载已有数据
    seq_data = np.load(seq_data_path)
    X_seq = seq_data['X_seq']
    y = seq_data['y']
    sequence_user_ids = seq_data['sequence_user_ids']
    print(f"✅ 从缓存加载时间序列数据（序列长度={SEQ_LENGTH}）")


print("🔄 步骤5：创建用户特征映射字典")
# 关键修复：先创建特征字典
from collections import defaultdict

# 将risk_features转换为字典格式
user_features_dict = risk_features.set_index('visitorid')[['total_events', 'cart_abandon_rate']].to_dict('index')

# 处理可能缺失的用户（设置默认值）
default_features = {'total_events': 0, 'cart_abandon_rate': 0.5}  # 示例默认值
user_id_to_features = defaultdict(lambda: default_features, user_features_dict)


print("🔄 步骤6：合并特征数据集")
# 修改特征提取方式：从字典中提取数值
X_risk = np.array([
    [user_id_to_features[uid]['total_events'],  # 提取第一个特征
     user_id_to_features[uid]['cart_abandon_rate']]  # 提取第二个特征
    for uid in sequence_user_ids
])

# 标准化（现在X_risk是二维数值数组）
X_risk = (X_risk - X_risk.mean(axis=0)) / X_risk.std(axis=0)

# 计算均值和标准差（这就是scalerParams需要的值）
mean_values = X_risk.mean(axis=0).tolist()
std_values = X_risk.std(axis=0).tolist()

# 保存为JSON
scaler_params = {
    "mean": mean_values,
    "std": std_values
}

import json
with open('scaler_params.json', 'w') as f:
    json.dump(scaler_params, f)

print("✅ 标准化参数已保存至 scaler_params.json")

# 验证维度
assert len(X_seq) == len(X_risk) == len(y), "数据维度不一致！"
print(f"✅ 合并完成 | 最终样本维度: {X_seq.shape}")
# 在步骤6后添加：
from sklearn.model_selection import train_test_split

# 划分数据集（必须在特征合并后进行）
X_train_seq, X_val_seq, X_train_risk, X_val_risk, y_train, y_val = train_test_split(
    X_seq, X_risk, y, test_size=0.2, random_state=42)
# %% [markdown]
# ## 2. 模型构建
print("🔄 步骤7：模型构建")
def build_model(seq_length, n_features, n_classes):
    # 序列输入分支
    seq_input = Input(shape=(seq_length,))
    embedding = Embedding(input_dim=3, output_dim=8)(seq_input)
    lstm_out = LSTM(64, return_sequences=False)(embedding)
    
    # 高危因子输入分支
    risk_input = Input(shape=(n_features,))
    
    # 合并分支
    combined = concatenate([lstm_out, risk_input])
    
    # 输出层
    output = Dense(n_classes, activation='softmax')(combined)
    
    model = Model(inputs=[seq_input, risk_input], outputs=output)
    model.compile(
        loss='sparse_categorical_crossentropy',
        optimizer='adam',
        metrics=[
            'accuracy',
            tf.keras.metrics.Precision(class_id=0, name='precision_view'),
            tf.keras.metrics.Recall(class_id=0, name='recall_view')
        ]
    )
    return model

model = build_model(seq_length=10, n_features=2, n_classes=3)
model.summary()
print("✅ 模型构建完成")

# %% [markdown]
# ## 3. 模型训练
print("🔄 步骤8：模型训练开始")
early_stop = EarlyStopping(monitor='val_loss', patience=3)

# 修改训练代码
history = model.fit(
    [X_train_seq, X_train_risk],
    y_train,
    validation_data=([X_val_seq, X_val_risk], y_val),
    epochs=20,
    batch_size=64,  # 调整为更小的批次大小
    callbacks=[early_stop],
    verbose=1
)


print("✅ 模型训练完成")

# 1. 保存为 SavedModel 格式（用于后续转换）
model.save("result_model.keras")

# 2. 转换为 ONNX 格式（确保已安装 tf2onnx）
import tf2onnx

# 1. 定义完整的输入签名
input_signature = [
    tf.TensorSpec(shape=(None, 10), dtype=tf.int32, name="sequence_input"),
    tf.TensorSpec(shape=(None, 2), dtype=tf.float32, name="risk_input")
]

# 执行转换
onnx_model, _ = tf2onnx.convert.from_keras(
    model,
    input_signature=input_signature,
    output_path="result_model.onnx"
)

print("✅ ONNX 转换完成")

# 创建并保存标签映射

# 假设 encoder 是 sklearn 的 LabelEncoder
label_to_index = {
    cls: int(label) for cls, label in zip(encoder.classes_, encoder.transform(encoder.classes_))
}

index_to_label = {
    int(label): cls for cls, label in zip(encoder.classes_, encoder.transform(encoder.classes_))
}

# 保存为 JSON
with open('label_to_index.json', 'w') as f:
    json.dump(label_to_index, f)

with open('index_to_label.json', 'w') as f:
    json.dump(index_to_label, f)

print("✅ 标签映射已保存")

# %% [markdown]
# ## 4. 结果评估
print("🔄 步骤9：绘制训练曲线")
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.legend()
plt.show()
print("✅ 训练曲线绘制完成")

print("🔄 步骤10：生成预测并输出分类报告")
y_pred = model.predict([X_val_seq, X_val_risk])
y_pred_class = np.argmax(y_pred, axis=1)  # 预测结果转换
y_true_class = y_val  # 直接使用原始标签

print(classification_report(y_true_class, y_pred_class, 
                            target_names=['view', 'addtocart', 'transaction']))
print("✅ 分类报告生成完成")

print("🎉 整个流程已全部完成！")
```
